Parte 1: Adquisici√≥n y An√°lisis Exploratorio (EDA)

```python
import pandas as pd

# 1. Carga del dataset original
try:
    df = pd.read_csv('datasets/train.csv')
    print(f"‚úÖ Archivo cargado correctamente. Registros totales: {len(df)}")
    
    # 2. An√°lisis de estructura jer√°rquica
    cat1_unicas = df['Cat1'].unique()
    cat2_unicas = df['Cat2'].unique()
    cat3_unicas = df['Cat3'].unique()

    print("\n--- CATEGOR√çAS NIVEL 1 (Macro-nichos) ---")
    print(cat1_unicas)

    print("\n--- CATEGOR√çAS NIVEL 2 (Sub-nichos) ---")
    print(cat2_unicas)

    print("\n--- CATEGOR√çAS NIVEL 3 (Micro-nichos) ---")
    print(cat3_unicas)

    print(f"\nResumen: {len(cat1_unicas)} en Cat1, {len(cat2_unicas)} en Cat2 y {len(cat3_unicas)} en Cat3.")

except Exception as e:
    print(f"‚ùå Error al cargar los datos: {e}")
```

Parte 2: Filtrado Estrat√©gico y Exportaci√≥n para Orange

```python
# 1. Definici√≥n de categor√≠as de inter√©s para Salud y Nutrici√≥n
categorias_nutricion = [
    'nutrition wellness', 'health care', 'beverages', 
    'breakfast foods', 'pantry staples', 'snack food', 
    'dairy eggs', 'baby food', 'feeding'
]

# 2. Aplicaci√≥n del doble filtro
mask_macro = df['Cat1'].isin(['health personal care', 'grocery gourmet food', 'baby products'])
mask_sub = df['Cat2'].isin(categorias_nutricion)
df_filtrado = df[mask_macro & mask_sub].copy()

# 3. Estructuraci√≥n con SCORE (Capa 3: Anal√≠tica de Datos)
# A√±adimos 'Score' a la selecci√≥n
corpus_orange = df_filtrado[['Cat1', 'Cat2', 'Cat3', 'Score', 'Text']].rename(
    columns={
        'Cat1': 'Macro_Nicho',
        'Cat2': 'Sub_Nicho',
        'Cat3': 'Detalle_Producto',
        'Score': 'Rating',
        'Text': 'Review'
    }
)

# 4. Limpieza y Exportaci√≥n
corpus_orange = corpus_orange.dropna(subset=['Review'])
corpus_orange.to_csv('health_master_corpus.csv', index=False)

print(f"---")
print(f"üéØ Corpus Maestro con Ratings generado.")
print(f"üìä Ahora Orange podr√° comparar el sentimiento con el Rating real.")
```
